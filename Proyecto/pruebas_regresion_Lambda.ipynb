```python
def evalua_polinomicos(L):
    resultados2 = []
    
    Vxp = ld.terminosPolinomicos(Vx,2)
    Exp = ld.terminosPolinomicos(Ex,2)
    Expn , scaler = ld.normalizar(Exp[:,1:])
    Expn = np.hstack([np.ones([Expn.shape[0],1]), Expn])

    Vxpn = scaler.transform(Vxp[:,1:])
    Vxpn = np.hstack([np.ones([Vxpn.shape[0],1]), Vxpn])
    
    
    for Lambda in L:
            
        precision, asignacion = rg.evaluar_validacion(Lambda,Expn,Ey ,Vxpn,Vy)
            
        print("Porcentaje de aciertos con lambda = {} : ".format(Lambda), precision, "%")
        resultados2.append(precision)
        print(Lambda, resultados2)
            
    plt.plot(L, resultados2)
```

```python
from scipy.spatial.distance import cosine

_ , _ , F = ld.cargarDatos()
E , P = train_test_split(F, test_size = 0.2 , random_state = random_state)
E = E.to_numpy()
P = P.to_numpy()

Ex , Ey = E[:,:-1] , E[:,-1]
Px , Py = P[:,:-1] , P[:,-1]

Exp = ld.terminosPolinomicos(Ex,2)
Expn, scaler = ld.normalizar(Exp[:,1:])
Expn = np.hstack([np.ones([Expn.shape[0],1]), Expn])

Pxp = ld.terminosPolinomicos(Px,2)
Pxpn = scaler.transform(Pxp[:,1:])
Pxpn = np.hstack([np.ones([Pxpn.shape[0],1]), Pxpn])

Theta = rg.oneVsAll(Expn,Ey,4,31.8)

r = []
for i in range(Pxpn.shape[0]):
    l = np.dot(Theta, Pxpn[i])
    m = max(l)
    i = np.where(l == m)
    r.append(i[0][0])

y2 = np.ravel(Py)
t = Px[np.where(r != y2)] # cuales se confunde
ty = Py[np.where(r != y2)]
```

 **Matriz de confusin**

```python
ConfMat = np.zeros([4,4])

for i in range(len(r)):
    c = int(Py[i])
    f = int(r[i])
    ConfMat[f,c] += 1
```

```python
import seaborn as sb
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.DataFrame(ConfMat , index = [i for i in range(4)] , columns = [i for i in range(4)])
plt.figure()
sb.heatmap(df , annot = True)
```

```python
metricas = []

for i in range(4) :
    precision = 0
    recall = 0
    
    for j in range(4):
        precision += ConfMat[j, i]
        recall += ConfMat[i, j]
    
    precision = ConfMat[i][i] / precision
    recall = ConfMat[i][i] / recall

    metricas.append((precision, recall))

print(metricas)
```
