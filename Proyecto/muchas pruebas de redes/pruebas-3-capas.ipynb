```python
from creador_redes import *
import loader as ld
import fun_basicas as fun
import matplotlib.pyplot as plt

random_state = 2714
INIT_EPSILON = 0.12

Ex, Ey, Vx, Vy, Px, Py = ld.carga_Numpy(random_state)
y_onehot = fun.one_hot(Ey, 4)

Ex2, scaler = ld.normalizar(Ex)
Vx2 = scaler.transform(Vx)
Px2 = scaler.transform(Px)
```

```python
redes = [63, 56, 49, 42, 35, 28, 21, 14, 7]
Lambdas = [0.001]

for r1 in redes:
    
    for r2 in redes:
        
        for r3 in redes:

            r = red_neuronal(3, [r1, r2, r3], 0.12, Ex.shape[1], 4)

            for L in Lambdas:

                iteraciones = [50, 100, 300, 500]
                validaciones = []
                pruebas = []

                for i in range(len(iteraciones)):
                    p1, p2 = r.entrenar(Ex, y_onehot, Vx, Vy, Px, Py, L, iteraciones[i], INIT_EPSILON)
                    validaciones.append(p1)
                    pruebas.append(p2)

                plt.figure()
                plt.title("red: {}".format([r1, r2, r3]))
                plt.plot(np.array(iteraciones), np.array(validaciones), label = 'validaciones' , Color = 'red')
                plt.plot(np.array(iteraciones), np.array(pruebas), label = 'pruebas' , Color = 'blue')
                plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from pandas.io.parsers import read_csv

d0 = read_csv('0.csv', header = None)
d1 = read_csv('1.csv', header = None)
d2 = read_csv('2.csv', header = None)
d3 = read_csv('3.csv', header = None)

df = pd.concat([d0,d1,d2,d3], axis = 0, ignore_index = True)

EyV , P = train_test_split(df, test_size = 0.2, random_state = random_state)
E = EyV.to_numpy()
P = P.to_numpy()

Ex = E[:,:-1]
Ey = np.ravel(E[:,-1])
    
Px = P[:,:-1]
Py = np.ravel(P[:,-1])

y_onehot = fun.one_hot(Ey, 4)

Ex2, scaler = ld.normalizar(Ex)
Px2 = scaler.transform(Px)
```
