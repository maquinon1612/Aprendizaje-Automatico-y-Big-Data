```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
```

```python

C = [10, 20, 30, 40, 50, 60, 70, 80]
S = [5, 10, 15 , 20, 25, 30, 35, 40]

resE = np.zeros((len(C), len(S)))
resP = np.zeros(resE.shape)

for c in range(len(C)):
    for s in range(len(S)):
        c1 = C[c]
        s1 = S[s]
        
        entrenamiento, prueba = svm.evaluar_probar(Ex2, Ey, Px2, Py, c1, s1)
        
        resP[c,s] = prueba
        resE[c,s] = entrenamiento

```

```python
CC , SS = np.meshgrid(C, S)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

superficie = ax.plot_surface(CC, SS, np.array(resP),cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

fig.colorbar(superficie, shrink=0.5, aspect=5)

plt.show()
```

```python
CC , SS = np.meshgrid(C, S)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

superficie = ax.plot_surface(CC, SS, np.array(resE),cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

fig.colorbar(superficie, shrink=0.5, aspect=5)

plt.show()
```

```python
from sklearn.metrics import plot_confusion_matrix
import sklearn.svm as skl
import numpy as np
```

```python
predictor = skl.SVC(kernel = 'rbf' , C = 30 , gamma = 1/(2*10**2))
predictor.fit(Ex2,Ey)
labels = np.array([0,1,2,3])

plot_confusion_matrix(predictor, Px2, Py, labels = labels)
```
