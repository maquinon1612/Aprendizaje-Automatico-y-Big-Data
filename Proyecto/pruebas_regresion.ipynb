```python
import loader as ld
import regresion as rg
import numpy as np
import matplotlib.pyplot as plt
random_state = 1706

Lambdas = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
```

```python
Ex, Ey, Vx, Vy, Px, Py = ld.carga_Numpy(random_state)
```

```python
total = Ex.shape[0] + Vx.shape[0] + Px.shape[0]
print(Ex.shape[0]/total * 100)
print(Vx.shape[0]/total * 100)
print(Px.shape[0]/total * 100)

print(Ex.shape[0])
```

```python
resultados = []
ejemplos = np.linspace(1, Ex.shape[0], 150, endpoint = False, dtype = int)

Vxn , esc2 = ld.normalizar(Vx)
Vxn = np.hstack([np.ones([Vxn.shape[0],1]), Vxn])

for i in range(ejemplos.shape[0]):

    Exn, esc1 = ld.normalizar(Ex[:ejemplos[i], :])

    ones = np.ones([ejemplos[i],1])
    Exn = np.hstack([ones, Exn])

    precision, asignacion = rg.evaluar_validacion(0,Exn,Ey[:ejemplos[i]],Vxn,Vy)
    resultados.append(precision)
```

```python
plt.plot(ejemplos, resultados)
```
